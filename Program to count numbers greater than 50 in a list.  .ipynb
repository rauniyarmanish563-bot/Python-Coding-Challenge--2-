```python
# Function to count numbers > 50
def count_greater(nums):
    count = 0

    # Loop through list
    for num in nums:
        if num > 50:
            count += 1

    return count

# Example input
nums = [10, 60, 30, 80]

# Calling function
print("Count:", count_greater(nums))
```
