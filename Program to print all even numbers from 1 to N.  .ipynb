```python
# Function to print even numbers
def print_even(n):
    # Loop from 1 to n
    for i in range(1, n+1):
        if i % 2 == 0:
            print(i, end=" ")

# Taking input
n = int(input("Enter number: "))

# Calling function
print_even(n)
```
