```python
# Function to check palindrome
def check_palindrome(num):
    original = num
    rev = 0

    # Reverse number
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    # Compare original and reversed
    if original == rev:
        return "Palindrome"
    else:
        return "Not Palindrome"

# Taking input
num = int(input("Enter number: "))

# Calling function
print(check_palindrome(num))
```
