```python
# Function to find largest number
def find_largest(a, b, c):
    # Comparing numbers
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Taking input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Calling function
print("Largest number is:", find_largest(a, b, c))
```
