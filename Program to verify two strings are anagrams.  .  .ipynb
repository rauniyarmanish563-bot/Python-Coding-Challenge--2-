```python
# Function to check anagram
def is_anagram(s1, s2):
    # Sort both strings and compare
    if sorted(s1) == sorted(s2):
        return "True"
    else:
        return "False"

# Input
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Output
print(is_anagram(str1, str2))
```
