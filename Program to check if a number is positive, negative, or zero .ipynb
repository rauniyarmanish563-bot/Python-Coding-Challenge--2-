```python
# Function to check number type
def check_number(num):
    # Conditions
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

# Taking input
num = int(input("Enter number: "))

# Calling function
print(check_number(num))
```
