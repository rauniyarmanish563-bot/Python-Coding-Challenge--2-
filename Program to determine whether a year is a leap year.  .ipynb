```python
# Function to check leap year
def check_leap_year(year):
    # Leap year condition
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return "Leap Year"
    else:
        return "Not a Leap Year"

# Taking input
year = int(input("Enter year: "))

# Calling function
print(check_leap_year(year))
```
