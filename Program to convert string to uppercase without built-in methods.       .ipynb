```python
# Function to convert to uppercase
def to_upper(s):
    result = ""

    for ch in s:
        # Check lowercase letters
        if 'a' <= ch <= 'z':
            result += chr(ord(ch) - 32)
        else:
            result += ch

    return result

# Input
text = input("Enter string: ")

# Output
print(to_upper(text))
```
