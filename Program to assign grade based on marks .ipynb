```python
# Function to assign grade
def assign_grade(marks):
    # Grade conditions
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 40:
        return "D"
    else:
        return "Fail"

# Taking input
marks = int(input("Enter marks: "))

# Calling function
print("Grade:", assign_grade(marks))
```
