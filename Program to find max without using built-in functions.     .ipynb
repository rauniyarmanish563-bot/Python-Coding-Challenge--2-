```python
# Function to find maximum manually
def find_max(nums):
    max_val = nums[0]   # assume first element is max

    # Loop through list
    for num in nums:
        if num > max_val:
            max_val = num   # update max

    return max_val

# Taking input (comma separated)
data = input("Enter numbers separated by comma: ")

# Convert string to list using comma
nums = list(map(int, data.split(",")))

# Calling function
print("Maximum value:", find_max(nums))
```
