```python
# Function to determine ticket price
def ticket_price(day):
    # Convert input to lowercase
    day = day.lower()

    # Check day
    if day == "sunday":
        return 200
    else:
        return 100  # normal price for other days

# Taking input
day = input("Enter day: ")

# Calling function
print("Ticket Price:", ticket_price(day))
```
