```python
# Function to calculate factorial
def factorial(n):
    fact = 1

    # Loop from 1 to n
    for i in range(1, n+1):
        fact *= i   # multiply

    return fact

# Taking input
num = int(input("Enter number: "))

# Calling function
print("Factorial:", factorial(num))
```
