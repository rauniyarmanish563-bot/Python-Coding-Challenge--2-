```python
# Function to count vowels
def count_vowels(s):
    count = 0

    # Loop through string
    for ch in s.lower():
        if ch in "aeiou":
            count += 1

    return count

# Input
text = input("Enter string: ")

# Output
print("Vowel count:", count_vowels(text))
```
