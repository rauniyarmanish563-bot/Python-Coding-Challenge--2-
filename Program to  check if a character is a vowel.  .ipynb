```python

# Function to check vowel
def check_vowel(ch):
    # Convert to lowercase
    ch = ch.lower()

    # Check vowel
    if ch in ['a', 'e', 'i', 'o', 'u']:
        return "Vowel"
    else:
        return "Not a Vowel"

# Taking input
char = input("Enter a character: ")

# Calling function
print(check_vowel(char))
```
