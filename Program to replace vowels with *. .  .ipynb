```python
# Function to replace vowels
def replace_vowels(s):
    result = ""

    for ch in s:
        if ch.lower() in "aeiou":
            result += "*"
        else:
            result += ch

    return result

# Input
text = input("Enter string: ")

# Output
print(result := replace_vowels(text))
```
