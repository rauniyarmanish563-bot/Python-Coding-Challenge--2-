```python
# Function to count character occurrences
def count_char(s, ch):
    count = 0

    for c in s:
        if c == ch:
            count += 1

    return count

# Input
text = input("Enter string: ")
char = input("Enter character: ")

# Output
print("Count:", count_char(text, char))
```
