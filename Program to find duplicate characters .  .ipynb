```python
# Function to find duplicates
def find_duplicates(s):
    seen = []
    duplicates = []

    for ch in s:
        if ch not in seen:
            seen.append(ch)
        elif ch not in duplicates:
            duplicates.append(ch)

    return duplicates

# Input
text = input("Enter string: ")

# Output
print("Duplicates:", " ".join(find_duplicates(text)))
```
