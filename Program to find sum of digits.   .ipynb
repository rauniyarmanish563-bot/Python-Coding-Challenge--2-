```python
# Function to find sum of digits
def sum_of_digits(num):
    total = 0

    # Loop until number becomes 0
    while num > 0:
        digit = num % 10
        total += digit
        num = num // 10

    return total

# Taking input
num = int(input("Enter number: "))

# Calling function
print("Sum of digits:", sum_of_digits(num))
```
