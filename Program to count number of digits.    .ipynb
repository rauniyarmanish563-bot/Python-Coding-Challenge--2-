```python
# Function to count digits
def count_digits(num):
    count = 0

    # Loop until number becomes 0
    while num > 0:
        count += 1
        num = num // 10

    return count

# Taking input
num = int(input("Enter number: "))

# Calling function
print("Number of digits:", count_digits(num))
```
