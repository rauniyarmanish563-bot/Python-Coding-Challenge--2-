```python
# Function to calculate total sales
def total_sales(sales_list):
    total = 0

    # Loop through list and add each value
    for sale in sales_list:
        total += sale

    return total

# Taking input from user
sales = []

print("Enter sales for 7 days:")

# Loop to take 7 inputs
for i in range(7):
    value = int(input(f"Day {i+1}: "))
    sales.append(value)   # add value to list

# Calling function
print("Total Sales:", total_sales(sales))
```
